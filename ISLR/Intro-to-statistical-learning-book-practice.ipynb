```r
#?rnorm

# set a seed for random variables (So the random numbers won't change on next run)
set.seed(6789)

# create a Normal Random Variables with sample-size of 10
A = rnorm(10)
B = rnorm(10, mean=50, sd=0.1)

print(A)
print(B)
```

```r
# ?mean
# ?var
# ?sd

set.seed(1)

A = rnorm(10, mean=50, sd=2)
m = mean(A)
d = sd(A)
v = var(A)

print(A)
print(m)
print(d)
print(v)
```

## Plot graphics

```r
# Create a Scatter Plot
x = rnorm(50)
y = rnorm(50)

plot(x, y)
```

```r
x = rnorm(50)
y = rnorm(50)

# save plot to jpeg format
jpeg('./plot.jpg') #create new file
    plot(x,y)
dev.off() #save data to file

# save plot to PDF format
pdf('./plot.pdf')
plot(x, y, col="green")
dev.off()
```

```r
# Create a [Contour Plot]
x = seq(-pi,pi,length=50)
y = x

f = outer(x,y,function(x,y)cos(y)/(1+x^2))
contour(x,y,f)

#contour(x,y,f,nlevels=45,add=T)
        
fa = (f-t(f))/2
contour(x,y,fa,nlevels=15)
```

```r
# ?image
# ?persp

# Create a heatmap
print('heatmap')
image(x,y,fa)

print('3D')
persp(x,y,fa)

print('3D')
persp(x,y,fa,theta=30)

print('3D')
persp(x,y,fa,theta=30,phi=20)

print('3D')
persp(x,y,fa,theta=30,phi=70) 

print('3D')
persp(x,y,fa,theta=30,phi=40)
```
